```python
ratings = data_final.groupby(['Aggregate rating','Rating color','Rating text' ]).size().reset_index().rename(columns={0:'Rating_count'})
```

```python
ratings
```

```python
sns.barplot(x='Aggregate rating',y='Rating_count',data=ratings,hue='Rating color',palette=['White','Red','Orange','Yellow','Green','Green'])
```

```python
k = data_final['Aggregate rating'].value_counts().reset_index().rename(columns={'index':'rating','Aggregate rating':'counts'})
```
